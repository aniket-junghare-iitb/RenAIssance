```python
import os
import shutil

txt_file = 'Dataset/partitions/train.txt'
image_folder = 'Dataset/images'            # Folder with all images
target_folder = 'Working_dataset/train'               # Destination folder

image_extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.tif', '.tiff']


os.makedirs(target_folder, exist_ok=True)


with open(txt_file, 'r') as f:
    image_names = [line.strip() for line in f if line.strip()]


for base_name in image_names:
    found = False
    for ext in image_extensions:
        full_name = base_name + ext
        source_path = os.path.join(image_folder, full_name)
        target_path = os.path.join(target_folder, full_name)

        if os.path.exists(source_path):
            shutil.move(source_path, target_path)
            found = True
            break

    if not found:
        print(f"Image not found: {base_name} (tried extensions: {image_extensions})")
```

```python
train_list_file = 'Dataset/partitions/train.txt'
transcriptions_file = 'Dataset/text/transcriptions.txt'
output_folder = 'Working_dataset/train_transcriptions'


os.makedirs(output_folder, exist_ok=True)


transcription_dict = {}
with open(transcriptions_file, 'r', encoding='utf-8') as f:
    for line in f:
        if line.strip():
            parts = line.strip().split(maxsplit=1)
            if len(parts) == 2:
                image_name, transcription = parts
                transcription_dict[image_name] = transcription


with open(train_list_file, 'r', encoding='utf-8') as f:
    train_image_names = [line.strip() for line in f if line.strip()]


for image_name in train_image_names:
    if image_name in transcription_dict:
        output_file = os.path.join(output_folder, image_name + '.txt')
        with open(output_file, 'w', encoding='utf-8') as out_f:
            out_f.write(transcription_dict[image_name])
    else:
        print(f"Transcription not found for: {image_name}")
```
